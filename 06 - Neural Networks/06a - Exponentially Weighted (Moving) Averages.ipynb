# Exponentially Weighted (Moving) Averages

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.style.use('ggplot')

plt.rcParams['figure.figsize'] = (12.0, 10.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'

# Fix the seed of the random number 
# generator so that your results will match ours
np.random.seed(1)

%load_ext autoreload
%autoreload 2
```

```python
x = np.arange(0,2*np.pi,0.01)
y = np.cos(x)+np.random.randn(x.shape[0])*0.4+1
plt.scatter(x,y)
```

```python
beta_list = np.arange(0.1,1.0,0.1)
pctr = 1
for beta in beta_list:
    v = [0]
    for i in range(1,len(x)+1):
        v.append(beta*v[i-1] + (1-beta)*y[i-1])
    plt.subplot(3,3,pctr)
    plt.scatter(x,y)
    plt.plot(x,v[1:],'b')
    plt.title(r'$\beta = '+str(beta)+'$')
    pctr+=1
```

```python
beta = 0.9
v = [0]
v_corrected = []
for i in range(1,len(x)+1):
    v.append((beta*v[i-1] + (1-beta)*y[i-1]))
    v_corrected.append(v[i] / (1-beta**i))
plt.plot(x,y)
plt.plot(x,v_corrected,'b')
plt.title(r'$\beta = '+str(beta)+'$')
```
